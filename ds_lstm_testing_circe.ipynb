```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding,Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import utils
```

```python
text = open("circe.txt", encoding="utf-8").read().splitlines()
```

```python
text
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)
```

```python
vocab_size = len(tokenizer.word_index)+1
vocab_size
```

```python
tokenizer.texts_to_sequences([text[0]])[0]
```

```python
input_sequence= []
for sent in text:
    sequence = tokenizer.texts_to_sequences([sent])[0]
    for j in range (1,len(sequence)):
        n_gram = sequence[:j+1]
        input_sequence.append(n_gram)
```

```python
input_sequence
```

```python
max_len= max([len(i)for i in input_sequence])
input_sequence = pad_sequences(input_sequence ,maxlen=max_len)
max_len
```

```python
inputs = input_sequence[:,:-1]
labels = input_sequence[:,-1]
labels = utils.to_categorical(labels,num_classes=vocab_size)
labels
```

```python

model = Sequential()
model.add(Embedding(vocab_size, 100, input_shape=(max_len-1,)))
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(128))
model.add(Dense(vocab_size, activation="softmax"))

# Compile

# Now summary will work
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=['accuracy'])
hestory=model.fit(inputs,labels,epochs=150)
```

```python
new_word = "He is half of my soul"
next_word = 100
for i in range(next_word):
    tokens = tokenizer.texts_to_sequences([new_word])
    padded = pad_sequences(tokens,maxlen=max_len-1)
    preds = model.predict(padded)
    prediction = np.argmax(preds,axis=1) 

    for word , index in tokenizer.word_index.items():
        if index == prediction:

            output_word = word
            break
    new_text = new_word+" "+output_word
print(new_text)
```
